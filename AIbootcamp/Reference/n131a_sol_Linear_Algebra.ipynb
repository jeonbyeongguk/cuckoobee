### **2-1. Norm**

- ì£¼ì–´ì§„ ê°’ $x_0, x_1, ... x_n$ ë¥¼ elementë¡œ ê°–ëŠ” $\vec v$ì— ëŒ€í•˜ì—¬ 
  - $L_1$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤. 
    > $|\vec v| = |x_0| + |x_1| + ... + |x_n|$ <br>
  - $L_2$ Norm ì˜ ì •ì˜ëŠ” ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.
    > $||\vec v|| = \sqrt {{x_0}^2 + {x_1}^2 + ... + {x_n}^2}$
- ì´ë¥¼ ì°¸ì¡°í•˜ì—¬ ì£¼ì–´ì§„ ë²¡í„°ì— ëŒ€í•´ì„œ $L_2$ Norm ì˜ ê°’ê³¼ $L_1$ Norm ì˜ ê°’ì˜ ì°¨ì˜ ì ˆëŒ€ê°’ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ë§Œë“œì„¸ìš”.

##### **Q3. $\vec v = [3, -1, 8, 9]$ì¼ ë•Œ, NormDif($\vec v$)ì˜ ê²°ê³¼ê°’ì„ ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ì  ë‘˜ì§¸ ìžë¦¬ê¹Œì§€ ì“°ì„¸ìš”.**

```python
def NormDif(v):
    L1_norm = np.sum(np.abs(v))
    L2_norm = np.sqrt(np.sum(np.square(v)))
    
    return np.abs(L2_norm - L1_norm)
```

```python
v = [3, -1, 8, 9]

round(NormDif(v), 2)
```

### **2-2. Errors**

- 2ê°œì˜ ë²¡í„°ì™€ errorì˜ íƒ€ìž…ì„ ìž…ë ¥ë°›ì•„ ê·¸ ê²°ê³¼ê°’ì„ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ë¥¼ ìž‘ì„±í•˜ì„¸ìš”.
  - errorì˜ íƒ€ìž…ì€ `MSE`, `MAE` 2ê°€ì§€ìž…ë‹ˆë‹¤.
> ${\rm MSE} = {1 \over n} \sum_{i=1}^n ({\hat Y_i}-Y_i)^2$
>
> ${\rm MAE} = {1 \over n} \sum_{i=1}^n |{\hat Y_i}-Y_i|$
>
> where $\hat Y_i$ : ì˜ˆì¸¡ê°’(prediction), $Y_i$ : ì‹¤ì œê°’(target) 

##### **Q4. ì˜ˆì¸¡ê°’ $x = [2085, 1815, 2235, 1400, 2500, 1430, 3070, 2000, 1299, 1180]$ ì´ê³ , ì‹¤ì œê°’ $y=[2103, 2099, 1985, 1630, 2480, 1300, 3200, 1990, 1405, 1210]$ ì¼ ë•Œ, error í•¨ìˆ˜ë¥¼ í†µí•´ ì–»ì€ MSEì™€ MAE ê²°ê³¼ê°’ì„ (MSE ê°’, MAE ê°’) í˜•íƒœë¡œ ì“°ì„¸ìš”.**

```python
def error(x, y, type):
    if type == 'MSE':
        err = np.sum(np.square(x-y))/len(x)
    else:
        err = np.sum(np.abs(x-y))/len(x)
        
    return err
```

```python
import numpy as np

x = np.array([2085,1815,2235,1400,2500,1430,3070,2000,1299,1180])
y = np.array([2103,2099,1985,1630,2480,1300,3200,1990,1405,1210])

print('MSE ê°’:', error(x, y, 'MSE'))
print('MAE ê°’:', error(x, y, 'MAE'))
```

```python
error(x, y, 'MSE'), error(x, y, 'MAE')
```

## **Part 4. Cramer's rule**

- **Cramer's ruleì„ ì‚¬ìš©í•´ $x_1$, $x_2$, $x_3$ì˜ ê°’ì„ êµ¬í•˜ì„¸ìš”. (ë‹¨, ë°˜ì˜¬ë¦¼í•˜ì—¬ ì†Œìˆ˜ì  ë‘˜ì§¸ ìžë¦¬ê¹Œì§€ ì“°ì„¸ìš”.)**
> $x_1$ + $2x_3$ = 6
>
> $-3x_1$ + $4x_2$ + $6x_3$ = 30
>
> $-x_1$ $-2x_2$ + $3x_3$ = 8

```python
# í–‰ë ¬ A
A = np.array([[1,0,2],
              [-3,4,6],
              [-1,-2,3]])
# í–‰ë ¬ B
B = np.array([[6],[30],[8]])

# í–‰ë ¬ A_j 
A_1 = np.array([[6,0,2],
              [30,4,6],
              [8,-2,3]])
A_2 = np.array([[1,6,2],
              [-3,30,6],
              [-1,8,3]])
A_3 = np.array([[1,0,6],
              [-3,4,30],
              [-1,-2,8]])

# Cramer's rule
x1 = round(np.linalg.det(A_1)/np.linalg.det(A),2)
x2 = round(np.linalg.det(A_2)/np.linalg.det(A),2)
x3 = round(np.linalg.det(A_3)/np.linalg.det(A),2)

print('x1=',x1,'\nx2=',x2,'\nx3=',x3)
```
