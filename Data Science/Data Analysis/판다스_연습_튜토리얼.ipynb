### **Question 58**

Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라


- Unknown : N

- Less than $40K : a

- 40K - $60K : b

- 60K - $80K : c

- 80K - $120K : d

- $120K +’ : e

```python
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/BankChurnersUp.csv',index_col=0)
def ch(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':   
        return 'b'
    elif x =='$60K - $80K':    
        return 'c'
    elif x =='$80K - $120K':   
        return 'd'
    elif x =='$120K +' :     
        return 'e'

df['newIncome'] = df.Income_Category.apply(ch)
df.newIncome
```

년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

- 2021-05-14:24  -> 2021-05-15:00
- 날짜하루 추가
- 달 추가


```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df = df.rename(columns = {'(년-월-일:시)': 'date'})
df.head()
```

```python
df[df['date'].str[-2:] == '24'].head(3)
```
